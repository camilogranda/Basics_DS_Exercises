```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = complex(2,3)

plt.plot(np.real(z),np.imag(z),'ro')
plt.plot([0,np.real(z)],[0,np.imag(z)],'r')

plt.xlim(-4,4)
plt.ylim(-4,4)

axis= plt.gca()
plt.plot(axis.get_xlim(),[0,0],'k')
plt.plot([0,0],axis.get_xlim(),'k')

plt.grid()
plt.xlabel('real')
plt.ylabel('imag')


plt.show()
```

```python
z1 = complex(-3,1)
z2 = complex(-1,1)
summ1 = z1+z2
prod1 = z1*z2

plt.plot([0,np.real(z1)],[0,np.imag(z1)],color=(0,0.7,0.5),label='z1')
plt.plot([0,np.real(z2)],[0,np.imag(z2)],'m',label='z2')
plt.plot([0,np.real(summ1)],[0,np.imag(summ1)],color=(0.1,0.2,0.9),label='z1+z2')

plt.xlim(-5,5)
plt.ylim(-5,5)

axis= plt.gca()
plt.plot(axis.get_xlim(),[0,0],'k')
plt.plot([0,0],axis.get_xlim(),'k')

plt.grid()
plt.legend()
plt.xlabel('real')
plt.ylabel('imag')

plt.show()
```

```python
plt.plot([0,np.real(z1)],[0,np.imag(z1)],color=(0,0.7,0.5),label='z1')
plt.plot([0,np.real(z2)],[0,np.imag(z2)],'m',label='z2')
plt.plot([0,np.real(prod1)],[0,np.imag(prod1)],color=(1,0.6,0.5),label='z1*z2')

plt.xlim(-5,5)
plt.ylim(-5,5)

axis= plt.gca()
plt.plot(axis.get_xlim(),[0,0],'k')
plt.plot([0,0],axis.get_xlim(),'k')

plt.grid()
plt.legend()
plt.xlabel('real')
plt.ylabel('imag')

plt.show()
```
