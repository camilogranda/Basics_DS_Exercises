```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
M = np.random.randint(-5,6,(5,5))
M = M@M.T

eigvals,eigvecs = np.linalg.eig(M)

fig,ax = plt.subplots(1,3)

ax[0].imshow(M)
ax[0].set_title('The matrix')

ax[1].imshow(np.diag(eigvals))
ax[1].set_title('Eigenvalues')

ax[2].imshow(eigvecs)
ax[2].set_title('Eigenvectors')

plt.show()
```

```python
# Eigendecomposition!
Mv = M@eigvecs[:,2]
Lv = eigvals[2] * eigvecs[:,2]

print(Mv)
print(Lv)
```

```python
Mat = np.array([[1,2],[2,1]])
# Mat = Mat@Mat.T

eigvals,eigvecs = np.linalg.eig(Mat)

Lv = eigvals[0] * eigvecs[:,0]
Mv = Mat@eigvecs[:,0]

plt.plot([0,Lv[0]],[0,Lv[1]],label='$\\lambda v$')
plt.plot([0,Mv[0]],[0,Mv[1]],'--',label='Mv')

plt.axis('square')
plt.axis([-3,3,-3,3])
plt.legend()
plt.grid()
plt.show()
```

### Singular value decomposition

```python
ein = Image.open('Einstein.jpg')
ein = np.mean(ein,2)
print(np.shape(ein))

plt.imshow(ein)

plt.show()
```

```python
U,s,V = np.linalg.svd(ein)

fig,ax = plt.subplots(1,3)

ax[0].imshow(U)
ax[0].set_title('U')

ax[1].imshow(np.diag(np.log(s)))
ax[1].set_title('$\\sigma$')

ax[2].imshow(V)
ax[2].set_title('V')

plt.show()
```

```python
plt.plot(s,'ks-',markerfacecolor='w')
plt.xlim(-1,50)
plt.title('Eigenspectrum of Einstein')
plt.show()
```

```python
randorder = np.random.permutation(len(s))

S = np.zeros(np.shape(ein))
for i in range(6,len(s)):
    S[i,i] = s[i]

rein = U@S@V

plt.subplot(1,2,1)
plt.imshow(ein)
plt.axis('off')
plt.title('original')

plt.subplot(1,2,2)
plt.imshow(rein)
plt.axis('off')
plt.title('SVD version')

plt.show()
```
