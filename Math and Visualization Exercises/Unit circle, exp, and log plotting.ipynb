```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(0, 2*np.pi,100)

plt.plot(np.cos(x),np.sin(x),'k')
plt.plot([-1.1,1.1],[0,0],'--',color=[0.5,0.5,0.5])
plt.plot([0,0],[-1.1,1.1],'--',color=[0.5,0.5,0.5])

angle = np.pi/2
plt.plot([0,np.cos(angle)],[0,np.sin(angle)],'r')
plt.axis('square')
plt.axis([-1.1,1.1,-1.1,1.1])
plt.xlabel('cos(x)')
plt.ylabel('sin(x)')

plt.show()
```

```python
x = np.linspace(-3,1.83)
k = np.exp(x)

for i in range(0,len(x)):
    plt.plot([0,np.cos(k)[i]],[0,np.sin(k)[i]])

plt.axis('square')
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.axis('off')

plt.show()
```

```python
x = np.linspace(-3,3,100)

plt.plot(x,np.exp(x))
plt.xlabel('x')
plt.ylabel('$e^x$')

plt.show()

plt.plot(x,np.log(x))
plt.xlabel('x')
plt.ylabel('$log(x)$')

plt.show()
```

```python
x = np.linspace(0.0001,10,10)

expx = np.exp(x)
logx = np.log(x)

y1 = np.log( expx )
y2 = np.exp( logx )

plt.plot(y1,'r',label='$\\log(e^x)$')
plt.plot(y2,'co',label='$e^{(\\log(x))}$')

plt.legend()

plt.show()
```
