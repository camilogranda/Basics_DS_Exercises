```python
import sympy as sym
import sympy.plotting.plot as symplot
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sym.symbols('x')

y = x**2

p = sym.plotting.plot(y,show=False)
p.xlim = [0,50]
p[0].line_color = 'm'
p.title = 'This is a nice plot'
p.show()
```

```python
x,a = sym.symbols('x,a')

expr = a/x

n = 1

p = symplot(expr.subs(a,n),(x,-5,5),show=False)
p[0].label = '$y=%s$' %sym.latex(expr.subs(a,n))

p2 = symplot(expr.subs(a,2),show=False)
p.extend(p2)
p[1].label = '$y=%s$' %sym.latex(expr.subs(a,2))

p[0].line_color = 'r'
p.ylim = [-5,5]
p.legend = True
p.show()
print(p)
```

```python
from matplotlib.ticker import FormatStrFormatter

x,a = sym.symbols('x,a')

y = a/(x**2-a)

p = None
p = symplot(y.subs(a,1),(x,-5,5),show=False)
p[0].label = '$%s$' %sym.latex(y.subs(a,1))

for i in range(2,5):
    p.extend(symplot(y.subs(a,i),(x,-5,5),show=False))
    p[i-1].label = '$%s$' %sym.latex(y.subs(a,i))
    p[i-1].line_color = list(np.random.rand(3))

p.ylim = [-10,20]
p.legend = True

plt.savefig('ThisIsANiceFigure1.png')
p.show()
```
