```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
a = 3
b = 4
c = np.sqrt( a**2 + b**2 )

plt.plot([0,a],[0,0],'k',linewidth='2')
plt.plot([0,0],[0,b],'k',linewidth='2')
plt.plot([0,a],[b,0],'k',linewidth='2')

plt.plot([.3,.3],[0,.3],'k',linewidth='1')
plt.plot([0,.3],[.3,.3],'k',linewidth='1')

plt.text(a/3,.3,'a=' + str(a),fontsize=12)
plt.text(.2,b/2.7,'b=' + str(b),fontsize=12)
plt.text(a/1.7,b/2,'c=' + str(c),fontsize=12)

plt.axis('square')
plt.axis('off')
axlim = np.max((a,b)) + .5
plt.axis([-.5,axlim,-.5,axlim])

plt.show()
```

### Pythagorean Theorem

```python
z = complex(3,4)

magnitude1 = np.sqrt(np.real(z)**2 + np.imag(z)**2)
magnitude2 = np.abs(z)
print('Hypotenuse of z = %s'%magnitude1)
print('Hypotenuse of z = %s'%magnitude2)

k = np.arctan2(np.imag(z),np.real(z))
print('Inverse tangent of z = %s'%k)
print('Inverse tangent of z = %s'%np.angle(z))


plt.plot([0,np.real(z)],[0,np.imag(z)])

plt.xlim(-4,4)
plt.ylim(-4,4)

axis = plt.gca()
plt.plot(axis.get_xlim(),[0,0],'k')
plt.plot([0,0],axis.get_xlim(),'k')

plt.show()
```

```python
x = np.linspace(0,6*np.pi,100)

plt.plot(np.sin(x),'r',label='$\\sin(\\theta)$')
plt.plot(np.cos(x),'b',label='$\\cos(\\theta)$')

plt.xlabel('rad.')
plt.ylabel('function value')
plt.legend()
plt.show()
```

```python
theta = np.linspace(0,4*np.pi,100)

y1 = np.sin(theta + np.cos(theta))
y2 = np.cos(theta + np.sin(theta))

plt.plot(y1,'g')
plt.plot(y2,'k')
plt.legend(['sin(x + cos(x))','cos(x + sin(x))'])

plt.show()
```

```python
x = np.linspace(0,2*np.pi,100)

thsym = sym.pi*7/6
theta = float(thsym)

# plot the circle
plt.plot(np.cos(x),np.sin(x),color=[.5,.5,.5])
plt.plot([0,0],[1.1,-1.1],'--',color=[.5,.5,.5])
plt.plot([1.1,-1.1],[0,0],'--',color=[.5,.5,.5])

# plot the arc
plt.plot(np.cos(np.linspace(0,theta)),np.sin(np.linspace(0,theta)),'k',linewidth=3)
plt.plot(np.cos(theta),np.sin(theta),'ko')
plt.plot([np.cos(theta),0],[np.sin(theta),0],'k:')

# plt the separate sine and cosine components
plt.plot([0,0],[0,np.sin(theta)],'r',linewidth=3,label=('sin part'))
plt.plot([np.cos(theta),0],[np.sin(theta),np.sin(theta)],'r:')
plt.plot([0,np.cos(theta)],[0,0],'g',linewidth=3,label=('cos part'))
plt.plot([np.cos(theta),np.cos(theta)],[np.sin(theta),0],'g:')

plt.axis('square')
plt.axis([-1.1,1.1,-1.1,1.1])
plt.title('$\\theta = %s $'%sym.latex(thsym))
plt.legend()

plt.show()
```
