```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[2,1,-1],[3,4,2],[1,-5,-2]])
y = np.array([1,13,0])

beta1 = np.linalg.inv(X.T@X) @ (X.T@y)
beta2 = np.linalg.solve(X,y)
print(beta1)
print(beta2)
```

```python
X1 = np.array([[3,-1],[-1,1]])
y1 = np.array([6,2])

X2 = np.array([[3,-1],[1.5,-.5]])
y2 = np.array([6,3])

X3 = np.array([[3,-1],[1.5,-.5]])
y3 = np.array([6,2])

print(X1)
```

```python
xlim = np.array([0,10])

#convert to y=mx+b format

yy1 = -X1[0,0]/X1[0,1]*xlim + y1[0]/X1[0,1]
yy2 = -X1[1,0]/X1[1,1]*xlim + y1[1]/X1[1,1]

plt.plot(xlim,yy1,label='eq1')
plt.plot(xlim,yy2,label='eq2')

#solve the systems
beta1 = np.linalg.solve(X1,y1)
print(beta1)

plt.plot(beta1[0],beta1[1],'ko',markersize=6,label='solution')

plt.legend()
plt.show()
```
