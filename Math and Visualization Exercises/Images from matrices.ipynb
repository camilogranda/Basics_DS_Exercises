```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = [[1,2],[1,4]]

plt.imshow(A)

plt.xticks([0,1])
plt.yticks([0,1])
```

```python
A = np.zeros([10,14])
print(np.shape(A))

for i in range(0,np.shape(A)[0]):
    for j in range(0,np.shape(A)[1]):
        A[i,j] = 3*i - 4*j
        
print(A)
plt.imshow(A)
plt.plot([0,6],[8,2],'k',linewidth=4)

for i in range(0,np.shape(A)[0]):
    for j in range(0,np.shape(A)[1]):
        plt.text(j,i,int(A[i,j]),horizontalalignment='center',verticalalignment='center')
        
plt.set_cmap('Spectral')

plt.show()
```

```python
N = 10
Ajz = np.zeros([N,N])

for i in range(0,np.shape(Ajz)[0]):
    for j in range(0,np.shape(Ajz)[1]):
        Ajz[i,j] = (-1)**(i+j)

plt.imshow(Ajz)      
plt.set_cmap('binary')
plt.tick_params(labelleft=False,labelbottom=False)
plt.show()
print(Ajz)
```
