```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = [1,2,2,2,2,2,3,4,5,6,6,6,6,7,8,9,8,6,3]

# plt.hist(data)
# plt.xlabel('values')
# plt.ylabel('Numbes of appearances (counts)')
# plt.show()

y,x = np.histogram(data,bins=np.arange(1,11))
yP = y/np.max(y) # y probability
print(x)
print(y)

plt.bar(x[:-1],yP)
plt.xlabel('Value')
plt.ylabel('Proportion (norm.)')
plt.xticks(np.arange(1,10))
plt.show()
```

```python
x = np.arange(-3,4)
y = x**2

yy,xx = np.histogram(y,bins=np.arange(1,8))
yyp = yy/np.sum(yy)
print(y)
print(yy)

plt.hist(y)
plt.xlabel('Values')
plt.ylabel('Counts')

plt.show()

plt.bar(xx[:-1],yyp)

plt.show()
```

```python
x = np.linspace(-3,3,1000)

fig,ax = plt.subplots(2,3)
fig.set_size_inches(10,6)

for i in range(3):
    
    y = x**(i+1)
    h = np.histogram(y,50)
    yP = h[0] / np.sum(h[0])

    ax[0,i].plot(x,y)
    ax[0,i].set_title('$y=x^%g$'%(i+1))
    
    ax[1,i].plot(h[1][:-1],yP)
```
