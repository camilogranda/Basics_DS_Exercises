```python
desired_mean = 15
desired_std = 4.3

nums = np.random.randn(1000) * desired_sd + desired_mean
nums = nums - np.mean(nums)
nums = nums / np.std(nums)
nums = nums * desired_std + desired_mean

plt.subplot(121)
plt.plot(nums,'g.')

plt.subplot(122)
plt.hist(nums)

plt.show()
print('The mean is: %s, and the SD is: %s'%(np.mean(nums),np.std(nums)))
```

## Convert angles function

```python
def angle_convert_plot():
    
    angle = eval(input('What angle to convert? '))
    unit = str(input('Which unit did you input? '))
    
    rad = np.deg2rad(angle)
    deg = np.rad2deg(angle)
        
    if unit[0].lower() == 'd':
        
        deg = angle
        rad = np.deg2rad(angle)
        
    elif unit[0].lower() == 'r':

        rad = angle
        deg = np.rad2deg(angle)
        
    else:
        raise ValueError('Unknown unit!')
    
    deg = deg%360
    rad = rad%(2*np.pi)
    
    plt.plot([0,0],[0,0],'ko',linewidth='3')
    plt.plot([0,1],[0,0],'k',linewidth='3')
    plt.plot([0,np.cos(rad)],[0,np.sin(rad)],'r',linewidth='4',alpha=0.65)
    
    plt.axis('square')
    plt.axis([-1,1,-1,1])
    plt.grid()
    plt.title('Angle is %g, or %g rad.'%(deg,rad))
    
    plt.show()   
```

```python
angle_convert_plot()
```
