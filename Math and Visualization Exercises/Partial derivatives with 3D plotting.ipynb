```python
import sympy as sym
from IPython.display import display,Math
```

```python
from sympy.abc import x,y
f = x**2 + x*y**2

display(Math('\\frac{\\partial f}{\\partial x} %s'%sym.latex(sym.diff(f,x))))
display(Math('\\frac{\\partial f}{\\partial y} %s'%sym.latex(sym.diff(f,y))))
```

```python
f = x**2 + x*y**2
diff_x = sym.diff(f,x)
diff_y = sym.diff(f,y)

sym.plotting.plot3d(f,(x,-3,3),(y,-3,3),title='$f(x,y)=%s$'%sym.latex(f))
sym.plotting.plot3d(diff_x,(x,-3,3),(y,-3,3),title='$\\frac{\\partial f}{\\partial x}=%s$'%sym.latex(diff_x))
sym.plotting.plot3d(diff_y,(x,-3,3),(y,-3,3),title='$\\frac{\\partial f}{\\partial y}=%s$'%sym.latex(diff_y))
```
