```python
x = np.linspace(0,2*np.pi,20)
fx = -(np.cos(x) + np.sqrt(x))

fxmax = np.max(fx)
fxmaxind = np.argmax(fx)
print(fxmax,fxmaxind)

plt.plot(x,fx,'bo-')
plt.plot(x[fxmaxind],fxmax,'rs')
print(x[fxmaxind])

plt.show()
```

```python
x = np.linspace(-2,2,51)
fx = x**3 + x**4

fxmin = np.min(fx)
fxminind = np.argmin(fx)

plt.plot(x,fx,'bs')
plt.plot(x[fxminind],fx[fxminind],'rs')

plt.xlim([-1.2,.4] )
plt.ylim([-.2,.2])

plt.show()
```

```python
x = np.linspace(0,12*np.pi,100)
fx = -np.cos(x) - np.sqrt(x)

peaks = find_peaks(fx)
print(peaks)
print(peaks[0])

plt.plot(x,fx)
plt.plot(x[peaks[0]],fx[peaks[0]],'s')

plt.show()
```

```python
x = np.linspace(-5,5,20)
fxeven = x**2
fxevenneg = (-x)**2

plt.plot(x,fxeven)
plt.plot(x,fxevenneg,'ro')
plt.show()
```

```python
fxodd = x**3
fxoddneg = (-x)**3
fxneg = -fxodd

plt.plot(x,fxodd)
plt.plot(x,fxoddneg,'ro')
plt.plot(x,fxneg,'k')

plt.show()
```

```python
theta = np.linspace(-2*np.pi,2*np.pi,39)
fxtheta  = np.cos(theta)
fxthetaN =  np.cos(-theta)

plt.plot(theta,fxtheta)
plt.plot(theta,fxthetaN,'ro')

plt.title('$cosine(\\theta)$')

plt.show()
```

```python
theta = np.linspace(-2*np.pi,2*np.pi,39)
sintheta  = np.sin(theta)
sinthetaN =  np.sin(-theta)

plt.plot(theta,sintheta)
plt.plot(theta,sinthetaN,'ro')

plt.title('$sin(\\theta)$')

plt.show()
```
