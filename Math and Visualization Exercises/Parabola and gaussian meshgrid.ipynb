```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
a = 1 # 1/4p
p = 1/4*a
h = 1
k = -1
n = 100

Focus = [h, k + p]
d = k - p

# x-axis points to evaluate the function
x = np.linspace(h-2,h+2,n)

# Create the function
y = a*(x-h)**2 + k

plt.plot(x,y,label='Parabola')
plt.plot(h,k,'ro',label='Vertex')
plt.plot(Focus[0],Focus[1],'go',label='Focus')
plt.plot(x[[0,-1]],[d,d],label='Directrix')

plt.grid()
plt.legend()
plt.axis([-1,3,-2,3])

plt.show()
```

```python
X,Y = np.meshgrid(range(0,10),range(0,14))

plt.subplot(121)
plt.pcolormesh(X,edgecolor='k',linewidth=1)

plt.subplot(122)
plt.pcolormesh(Y,edgecolor='k',linewidth=1)

plt.show()
```

```python

x = (np.linspace(-2,2,100))
X,Y = np.meshgrid(x,x)
s = 2
z = np.exp( -(X**2+Y**2)/s )

plt.imshow(z)
plt.axis('off')

plt.show()
```
