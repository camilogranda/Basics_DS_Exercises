```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numbers = np.arange(2,51)

a = []
c = []
b = []
for m in numbers:
    for n in numbers:
        a.append(m**2 - n**2)
        b.append(2*m*n)
        c.append(m**2 + n**2)
```

```python
fig,ax = plt.subplots(1,figsize=(10,7))
plt.scatter(a,b,c=c,marker='o',alpha=.4,edgecolor='g',linewidth=.3)
plt.axis('off')
plt.show()
```

```python
numbers = np.arange(2,51)

a = []
c = []
b = []
for m in numbers:
    for n in range(numbers[0],m):
        a.append(m**2 - n**2)
        b.append(2*m*n)
        c.append(m**2 + n**2)
```

```python
fig,ax = plt.subplots(1,figsize=(10,7))
plt.scatter(a,b,c=c,marker='o',alpha=.4,edgecolor='g',linewidth=.3)
plt.axis('off')
plt.show()
```

```python
pythcheck = np.array(a)**2 + np.array(b)**2 - np.array(c)**2

plt.plot(pythcheck,'o')
plt.show()
```
