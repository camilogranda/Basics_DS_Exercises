```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

```python
n = 4

for k in range(n):
    z = sym.exp( 2*sym.pi*sym.I*k/n )
    display(Math('(%s)^{%s} \\Rightarrow %s'%(sym.latex(z),n,z**n)))
```

```python
n = 10

for k in range(n):
    z = (np.exp( 2*np.pi*1j*k/n ))
    plt.plot([0,np.real(z)],[0,np.imag(z)])

x = np.linspace(0,2*np.pi)
plt.plot(np.cos(x),np.sin(x),'o',color='gray')

plt.axis('square')

plt.show()
```

```python
n = 199

color = np.linspace(0,.9,n)

for k in range(n):
    z = k*( np.exp( 2*np.pi*1j*k/n ))
    plt.plot([0,np.real(z)],[0,np.imag(z)],linewidth='3',color=[color[k],color[k],color[k]])
    
plt.axis('square')
plt.axis('off')

plt.show()
```

```python
n1 = 70
n2 = 100
a = np.log10(n1)
b = np.log10(n2)
n = 100

lins = np.linspace(n1,n2,n)
logs = np.logspace(a,b,n)

plt.plot(lins,label='linear')
plt.plot(logs,label='log')

plt.legend()

plt.show()
```

```python
a = 3
b = 4

res1 = np.log(a*b)
res2 = np.log(a)*np.log(b)
res3 = np.log(a)+np.log(b)

display(Math('\\log(%g\\times%g)= %g'%(a,b,res1)))
display(Math('\\log(%g)\\times \\log(%g)= %g'%(a,b,res2)))
display(Math('\\log(%g) + \\log(%g)= %g'%(a,b,res3)))

display(Math('\\log(%g\\times%g)= \\log(%g) + \\log(%g)'%(a,b,a,b)))
display(Math('\\log(%g\\times%g)\\neq \\log(%g)\\times \\log(%g)'%(a,b,a,b)))
```

```python
res1 = np.log(a/b)
res2 = np.log(a)/np.log(b)
res3 = np.log(a)-np.log(b)

display(Math('\\log\\left(\\frac{%g}{%g}\\right)= %g'%(a,b,res1)))
display(Math('\\log(%g)/\\log(%g)= %g'%(a,b,res2)))
display(Math('\\log(%g) - \\log(%g)= %g'%(a,b,res3)))

display(Math('\\log\\left(\\frac{%g}{%g}\\right)= \\log(%g) - \\log(%g)'%(a,b,a,b)))
display(Math('\\log\\left(\\frac{%g}{%g}\\right)\\neq \\frac{\\log(%g)}{\\log(%g)}'%(a,b,a,b)))
```

```python
a = 5
b = 3

rslt1 = np.log(a**b)
rslt2 = b*np.log(a)

display(Math('\\log(%g^%g)=%g'%(a,b,rslt1)))
display(Math('%g\\times\\log(%g)=%g'%(b,a,rslt2)))

display(Math('\\log(%g^%g)=%g\\times\\log(%g)'%(a,b,b,a)))
```
