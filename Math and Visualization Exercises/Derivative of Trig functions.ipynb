```python
import sympy as sym
from IPython.display import display,Math
```

```python
q = sym.symbols('q')
f = sym.cos(q)

for i in range(0,4):
    display(Math('\\frac{d}{dx}%s = %s'%(sym.latex(f),sym.latex(sym.diff(f)))))
    f = sym.diff(f)
```

```python
import sympy.plotting.plot as symplot

f = sym.cos(q)

for i in range(0,4):
    if i==0:
        p = symplot(f,show=False,label=f)
    else:
        p.extend(symplot(f,show=False,label=f))
    f = sym.diff(f)
    
p.legend = True
p.xlim = [-6,6]
p.show()
```

```python
import sympy.plotting.plot as symplot

x,a = sym.symbols('x,a')

f = sym.cos( x + sym.sin(x) ) + a
colors = 'brkm'

for ai in range(0,4):
    if ai == 0:
        p = symplot(f.subs(a,ai),show=False,label='a='+str(ai),line_color=colors[0])
    else:
        p.extend(symplot(f.subs(a,ai),show=False,label='a='+str(ai),line_color=colors[ai]))

p.legend = True
p.title = 'The functions'
p.show()

for ai in range(0,4):
    if ai == 0:
        p = symplot(sym.diff(f.subs(a,ai)),show=False,label='a='+str(ai))
    else:
        p.extend(symplot(sym.diff(f.subs(a,ai)),show=False,label='a='+str(ai)))
    
p.legend = True
p.title = 'The derivatives'
p.show()
```
