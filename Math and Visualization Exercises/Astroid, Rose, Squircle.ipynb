```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-6*np.pi,6*np.pi,500)

fg,ax = plt.subplots(3,3,figsize=(10,6))
row, col = np.indices((3,3))

print(row)
print(' ')
print(col)

for i in range(9):
    a = t**i
    
    x = a*np.cos(t)**3
    y = a*np.sin(t)**3
    
    r = row.ravel()[i]
    c = col.ravel()[i]
    
#     plt.figure()
    ax[r,c].plot(x,y,'k',linewidth=2)
    ax[r,c].axis('square')
    ax[r,c].axis('off')
plt.show
```

```python
k = 4.248

t = np.linspace(0,8*np.pi,1000)

x = np.cos(k*t) * np.cos(t)
y = np.cos(k*t) * np.sin(t)

plt.plot(x,y,'m',linewidth=3)
plt.axis('square')
plt.axis('off')
plt.show()
```

```python
t = np.linspace(0,8*np.pi,1000)

fig,ax = plt.subplots(3,3,figsize=(10,6))
row, col = np.indices((3,3))



for i in range(9):
    k = i/2
    
    x = np.cos(k*t) * np.cos(t)
    y = np.cos(k*t) * np.sin(t)
    
    r = row.ravel()[i]
    c = col.ravel()[i]
    
#     plt.figure()
    ax[r,c].plot(x,y,'k',linewidth=1)
    ax[r,c].set_title('k=%s'%k)
    ax[r,c].axis('square')
    ax[r,c].axis('off')
plt.show()
```

### Squircle

```python
x = np.linspace(-1,1,100)
y = (1-x**4)**(1/4)

plt.plot(x,y)
plt.plot(x,-y)
plt.axis('square')
plt.axis([-2,2,-2,2])
```
