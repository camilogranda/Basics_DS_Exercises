```python
import matplotlib.pyplot as plt
import sympy as sym
import numpy as np
from IPython.display import display, Math
```

```python
x = sym.symbols('x')
y = sym.sqrt(abs(x))
y

for i in range(-25,25):
    plt.plot(i,y.subs(x,i))
    plt.plot([i,0],[y.subs(x,i),0],linewidth='1.5')


plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
plt.plot([0,0],[0,2],'g')
plt.plot([2,2],[0,2],'m')
plt.plot([0,2],[0,0],'k')
plt.plot([0,2],[2,2],color=(1,0.3,0))

plt.axis('square')
plt.axis([-3, 5, -3, 5])
plt.title('A colorful square')
plt.show()
```

```python
#y = mx + b

x = [-5,5]
m = 2
b = 1

# y = [13,23]
# for i in range(0,len(x)):
#     y[i] = m*x[i] + b

y = m*np.array(x)+b

plt.plot(x,y,label=('y=%sx+%s'%(m,b)))

plt.axis('square')
plt.grid()
plt.legend()
plt.axis([-11,11,-11,11])

axis = plt.gca()
plt.plot(axis.get_xlim(),[0,0],'k:')
plt.plot([0,0],axis.get_ylim(),'k:')
plt.show()
```

```python
x = [-4.5,4.5]
m1 = 0.7
b1 = -2
y = m1*np.array(x)+b1

m2 = -1.25
b2 = 0.75
y2 = m2*np.array(x)+b2

plt.plot(x,y,label=('y=%sx+%s'%(m1,b1)))
plt.plot(x,y2,label=('y=%sx+%s'%(m2,b2)))

plt.legend()
plt.show()
```

```python
x = [-5,5]
m = [0.7,-1.25]
b = [-2,0.75]

for i in range(0,len(m)):
    y = m[i]*np.array(x)+b[i]
    plt.plot(x,y,label=('y=%s*%s+%s'%(m[i],x[i],b[i])))

plt.grid()
plt.legend()
axis = plt.gca()
plt.plot(axis.get_xlim(),[0,0],'k-.')
plt.plot([0,0],axis.get_ylim(),'k-.')
plt.show()

# for i in range(0,len(x)):
#     y[i] = (np.array(m*x+b))[i]
```

```python
x = range(-3,4)
y = np.zeros(len(x))

for i in range(0,len(x)):
    y[i] = 2 - x[i]**2
    
plt.plot(x,y,'s-')
plt.show()
```

```python
x = np.linspace(-3,4,20)
y = 2 + np.sqrt(abs(x))

plt.plot(x,y,'mo-.')
```

```python
x = np.linspace(-4,4,100)
exp = range(-1,4)

for i in exp:
    y = x**i
    label = display(Math('y=x^%s' %(i)))
    plt.plot(x,y,linewidth='4',label='$y=x^{%g}$'%(i))

plt.xlabel('x')
plt.ylabel('y')
plt.axis([-4,4,-20,20])
plt.legend()
plt.show()
```
