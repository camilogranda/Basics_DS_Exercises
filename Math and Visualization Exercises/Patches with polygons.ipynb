```python
import numpy as np
from matplotlib.patches import Polygon
import matplotlib.pyplot as plt
```

```python
from matplotlib.ticker import FormatStrFormatter
y = ([[1,1],[2,3],[3,1]])
p = Polygon(y,facecolor='c',alpha=0.5,edgecolor='k')

y1 = ([[1,3],[3,3],[2,1]])
p1 = Polygon(y1,facecolor='m',alpha=0.5,edgecolor='k')

fig,ax = plt.subplots()

ax.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
ax.xaxis.set_major_formatter(FormatStrFormatter('%.2f'))

ax.add_patch(p)
ax.add_patch(p1)

ax.set_xlim([0,4])
ax.set_ylim([0,4])

plt.show()
```

```python
x = range(-2,2)
y = np.zeros(len(x))

for i in range(0,len(x)):
    y[i] =  -x[i]**2
    
plt.plot(x,y,'o-')
plt.axis([-,2,2,-4,0])
plt.show()
```

```python
x = np.linspace(-2,2,100)
f = -x**2

pnts = np.vstack((x,f)).T
p = Polygon(pnts,facecolor='g',alpha=0.4)

y = ([[-0.5,-4],[-0.5,-2.5],[0.5,-2.5],[0.5,-4]])
p2 = Polygon(y,facecolor='k')



fg,ax = plt.subplots()

ax.add_patch(p)
ax.add_patch(p2)
plt.plot(x,f,'k')
plt.plot([-2,2],[-4,-4],'k')
plt.axis('off')
plt.show()
    
# yv
# plt.plot(x,y)
# plt.show()
# print(y)

# yp = ([[-2,-4],[0,yv],[2,-4]])
# p = Polygon(yp)
# fig,ax = plt.subplots()
# ax.add_patch(p)
# ax.set_xlim([-2.5,2.5])
# ax.set_ylim([-4.5,0.5])

```
