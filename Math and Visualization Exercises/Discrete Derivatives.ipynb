```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lst = [1,3,4,1,6]

print(sum(lst))
print(np.sum(lst))
np.cumsum(lst)

plt.plot(lst,'rd-.')
plt.plot(np.cumsum(lst),'bo:')
plt.legend(['sum','cumsum'])

plt.show()
```

```python
a = np.array([1,2,3,4,5])

suma = np.sum(a)
suma2 = np.sum(a**2)

ans1 = suma/suma2
ans2 = 1/suma

print('Option 1 = %s'%(ans1))
print('Option 2 = %s'%(ans2))
ans1 == ans2
```

```python
prda = np.product(a)
prda2 = np.product(a**2)

pans1 = prda/prda2

pans2 = 1/prda

print('Option 1 = %s'%(pans1))
print('Option 2 = %s'%(pans2))

pans1 == pans2
```

### Differences (Discrete Derivatives)

```python
v = np.arange(0,11)
print(v)
print(np.diff(v))
print(np.diff(v,2))
print(np.diff(np.diff(v)))

```

```python
x = np.linspace(-1.5,1.5,30)
f = x**2

dx = np.diff(x)
df = np.diff(f)

plt.plot(x,f,label='f',color='g')
plt.plot(x[:-1],df/dx,label='df/dx',color='k')

plt.axis([-2,2,-1,2])
plt.grid()
plt.legend()
plt.show()
```
