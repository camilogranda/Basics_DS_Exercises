```python
import sympy as sym
import numpy as np
from IPython.display import display,Math
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
```

```python
x = sym.symbols('x')

f = x

sym.integrate(f)
p = sym.plotting.plot(f,show=False)
p.xlim = [0,2]
p.ylim = [0,2]
p.show()
```

```python
f = x
sym.integrate(f,(x,0,1))
```

```python
f = x**3 / (x-2)
indf = sym.integrate(f)

p = sym.plotting.plot(f,show=False)
p.extend(sym.plotting.plot(indf,(x,2.1,10),show=False))

p.ylim = [-200,200]
p[0].label = '$f(x) = %s $' %(sym.latex(f))
p[1].label = '$\\int f(x) dx = %s $' %(sym.latex(indf))

p.legend = True
p.show()
```

### Fundamental Theorem of Calculus

```python
f = 2*x**3 + sym.sin(x)
diff_f = sym.diff(f)
ind_f = sym.integrate(f)
ind_diff = sym.integrate(diff_f)

display(Math('f(x) = %s'%sym.latex(f)))
display(Math('f\'(x) = %s'%sym.latex(diff_f)))
display(Math('\int f(x) = %s'%sym.latex(ind_f)))
display(Math('\int f\'(x) = %s'%sym.latex(ind_diff)))
```

### Area between two curves

```python
x = sym.symbols('x')

symf = x**2
symg = x

f = sym.lambdify(x,symf)
g = sym.lambdify(x,symg)

xx = np.linspace(-2,2,55)

# Add patch
xpatch = np.linspace(0,1,100)
ypatch = np.vstack((g(xpatch),f(xpatch))).T

fig,ax = plt.subplots()
ax.add_patch( Polygon(ypatch,facecolor='k',alpha=.6))

plt.plot(xx,f(xx))
plt.plot(xx,g(xx))

plt.legend(['$f(x)=%s$'%sym.latex(symf),'$g(x)=%s$'%sym.latex(symg)])
plt.xlim(-.25,1.5)
plt.ylim(-.5,1.5)

plt.show()
```

### Computing the area between two functions

The area between two functions is given by the formula $A=\int_{a}^{b} f(x)-g(x) dx$

In our example, $f(x) = x^2 \text { and } g(x) = x$

Therefore $A = \int_{a}^{b} (x-x^2)dx$

We will compute the area between the two crossing points, that is, where the two functions are equal. This is given by the two solutions to the equation 
$$x^2 = x$$

The two solutions are $x = 0$ and $x = 1.$ This gives us the definite integral of 
$$A=\int_{0}^{1}(x-x^2)dx$$

The final answer will be given in the code below.



```python
fg_intersect = sym.solve(symf-symg)
display(Math('f(x) \\text{ and } g(x) \\text { intersect at } x = %g \\text{ and } %g'%(fg_intersect[0],fg_intersect[1])))

#compute the area
A = sym.integrate(symg-symf,(x,fg_intersect[0],fg_intersect[1]))
display(Math('\\text{ The area between this two functions is } %s'%A))
```
