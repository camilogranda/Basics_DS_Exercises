```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = np.pi/4
m = 2.3

eul = m*np.exp(1j*k)
cis = m*( np.cos(k) + 1j*np.sin(k) ) # cis = cosine-imaginary-sine

print(eul)
print(cis)
```

```python
mag = np.abs(eul)
ang = np.angle(eul)

plt.polar([0,ang],[0,mag],'r')
plt.polar(k,m,'bo')

plt.show()
```

```python
def eulerFromCosSin():
    re = eval(input('input the cosine part: '))
    im = eval(input('input the sine part: '))
    
    m = np.sqrt(re**2 + im**2)
    k = np.arctan2(im,re)
    
    plt.polar([0,k],[0,m])
    
    plt.title('me$^{i\\phi}$, m=%g, $\\phi$= %g'%(m,k))
    plt.thetagrids([.68,48.37,98.67,129.78,228.5,287.8])
    plt.show()
    
```

```python
nvects = 200
ms = np.random.rand(nvects)
ks = np.random.rand(nvects)*2*np.pi

plt.axes(projection = 'polar')
for i in range(0,100):
    
    r = np.random.rand()
    
    if r<.4:
        clr = [1,.2,.7]
    elif r>.4 and r<.8:
        clr = [.7,.2,1]
    else:
        clr = [0,1,0]
        
    plt.plot([0,ks[i]],[0,ms[i]],color=clr)
    
plt.axis('off')
plt.show()
```
