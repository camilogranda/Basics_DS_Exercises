```python
import matplotlib.pyplot as plt
import sympy as sym
```

```python
x = 3
y = 5

plt.plot(x,y, 'b.')

plt.axis('square')

plt.axis([-6,6,-6,6])
plt.show()
```

```python
x = [2, 5, 7, -3, 4, -7, 1]
y = [4, -5, 2, 6, 4, -2, 8]

for i in range(len(x)):
    plt. plot(x[i], y[i], 'o', label='point %s'%(i+1))

plt.axis('square')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(4,2,'r+')

axis = plt.gca() #gca means 'get current axis'
ylim = axis.get_ylim()
axis.set_ylim(ylim[0], 2.345)
print(ylim)

plt.show()
```

```python
x = sym.symbols('x')
y = x**2 - 3 * x

for xi in range(-11,11):
    
    plt.plot(xi, y.subs(x, xi), 'o')

plt.grid()
plt.xlabel('x')
plt.ylabel('f(x) = %s'%y)
plt.show()
```
