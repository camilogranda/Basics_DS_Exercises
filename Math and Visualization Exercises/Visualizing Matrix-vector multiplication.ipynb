```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = np.array([[.5,1],
              [1,.5]])
v = np.array([1,-1])
# v = np.random.randn(2)

Mv = M@v
print(M)

plt.plot([0,v[0]],[0,v[1]],linewidth=3,label='v')
plt.plot([0,Mv[0]],[0,Mv[1]],'--',linewidth=3,label='Mv')

plt.axis('square')
plt.axis([-5,5,-5,5])
plt.grid()
plt.legend()

plt.show()
```

```python
M = np.array([[.5,1],
              [1,.5]])

vcomp = np.linspace(-2,2,40)

for a in vcomp:
    v = np.array([1,a])
    Mv = M@v
    plt.plot([0,v[0]],[0,v[1]],linewidth=2,color=[0,1-abs(a)/2,1-abs(a)/2],alpha=.8)
    plt.plot([0,Mv[0]],[0,Mv[1]],linewidth=2,color=[1-abs(a)/3,0,abs(a)/6])
    
    
plt.axis('square')
plt.axis('off')
plt.show()
```
