```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 2000

sign1 = np.sin(np.linspace(0,6*np.pi,n))
sign2 = np.random.rand(n)*2-1

combi = sign1 + sign2

fig,ax = plt.subplots(2,3,figsize=(10,4))

ax[0,0].plot(sign1)
ax[0,0].set_title('Sine wave')
ax[1,0].hist(sign1,100)

ax[0,1].plot(sign2,'.',alpha=.3)
ax[0,1].set_title('Uniform numbers')
ax[1,1].hist(sign2,100)

ax[0,2].plot(combi)
ax[0,2].set_title('Combined signal')
ax[1,2].hist(combi,100)

for i in np.matrix.flatten(ax):
    i.set_xticks([])
    i.set_yticks([])
    
plt.show()
```

```python
import seaborn as sns

sns.set()

sns.displot(combi,bins=100,kde=True)

plt.show()
```

```python
ns = np.arange(50,10000,100)
nbins = 50

allhist = np.zeros((len(ns),nbins))

for i in range(len(ns)):
    signal1 = np.sin(np.linspace(0,6*np.pi,ns[i]))
    signal2 = (np.random.rand(ns[i])*2-1)
    
    allhist[i,:],x = np.histogram(signal1+signal2,nbins,density=True)
    
plt.pcolor(x[:-1],ns,allhist,vmax=.6,cmap='viridis')
plt.xlabel('Data Value')
plt.ylabel('Points in the sample')
plt.colorbar()
plt.show()
```
