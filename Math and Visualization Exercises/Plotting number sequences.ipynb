```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xx = np.arange(1,81)

s1 = []
s2 = []
s3 = []
s4 = []
for a in xx:
    s1.append(a/(a+1))
    s2.append((a+1)/a)
    s3.append(a/sum(np.arange(1,a+1)))
    s4.append((-1)**a/np.sqrt(a))

fig,ax = plt.subplots(1,figsize=(8,6))
plt.plot(xx,s1,label='$\\frac{a}{a+1}$')
plt.plot(xx,s2,label='$\\frac{a+1}{a}$')
plt.plot(xx,s3,label='$\\frac{a}{\Sigma^{a}_{1}a}$')
plt.plot(xx,s4,label='$\\frac{-1^a}{\sqrt{a}}$')

plt.legend(loc='upper right',fontsize=16)
plt.show()
```

```python
xx = np.arange(1,81)

s1 = []
s2 = []
s3 = []
s4 = []
for a in xx:
    s1.append(a/(a+1))
    s2.append((a+1)/a)
    s3.append(sum(np.arange(1,a+1)/a))
    s4.append(np.sqrt(a)/(-1)**a)

fig,ax = plt.subplots(1,figsize=(8,6))
plt.plot(xx,s1,label='$\\frac{a}{a+1}$')
plt.plot(xx,s2,label='$\\frac{a+1}{a}$')
plt.plot(xx,s3,label='$\\frac{a}{\Sigma^{a}_{1}a}$')
plt.plot(xx,s4,label='$\\frac{-1^a}{\sqrt{a}}$')

plt.legend(loc='upper right',fontsize=16)
plt.show()
```
