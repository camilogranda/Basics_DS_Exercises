```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

### Role of variance on p-value and T-test

```python
## generate the data

# parameters
N = 20  # sample size
popMu = .5 # true population mean
std = np.linspace(.1,3,300)

# the null-hypothesis value
H0val = 0

data = np.zeros((N,len(std)))
print(np.shape(data))
t = np.zeros((len(std)))
p = np.zeros((len(std)))
print(len(std))
for i in range(len(std)):
    data[:,i]  = std[i]*np.random.randn(N) + popMu
    t[i],p[i] = stats.ttest_1samp(data[:,i],H0val)
    
# for j in range(len(std)):
#     t[j],p[j] = stats.ttest_1samp(data[:,j],H0val)
print(data)
plt.plot(std,t,'s',markerfacecolor='w',markersize=4.5)
plt.xlabel('Standard deviation')
plt.ylabel('T')
plt.show()

plt.plot(std,p,'s',markerfacecolor='w',markersize=4.5)
plt.xlabel('Standard deviation')
plt.ylabel('p-value')
plt.show()

plt.plot(t,p,'o',markerfacecolor='w',markersize=4.5)
plt.xlabel('T')
plt.ylabel('p-value')
plt.show()

# # let's see what the data look(s) like
# plt.plot(data,'ko',markerfacecolor='w',markersize=10)
# plt.xlabel('Data index')
# plt.ylabel('Data value')
# plt.show()
```
