```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

### Pearson's r vs Kendall Tau-b on ordinal data

```python
# print(np.ceil(7.5*np.random.rand(10)))

N = 100
d1 = np.random.randint(low=1,high=10,size=N)
noises = np.arange(-1,1,.01)
rslts = np.zeros((len(noises),2))

for idx,noise in enumerate(noises):
    d2 = np.ceil((d1+np.random.randint(low=0,high=10,size=N)*noise))
    rslts[idx,0] = stats.kendalltau(d1,d2)[0]
    rslts[idx,1] = stats.pearsonr(d1,d2)[0]

fig,ax = plt.subplots(figsize=(10,5))

ax.plot(noises,rslts.T[0],'d',label='Kendall Tau-b')
ax.set_xlabel('Noise factor')
ax.set_ylabel('Correlation coefficient')
ax.plot(noises,rslts.T[1],'d',label='Pearson r')

plt.legend()
plt.show()
```
