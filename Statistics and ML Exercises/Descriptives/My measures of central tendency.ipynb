```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
nbins = 8
data1 = np.random.randn(n)
dataexp = np.exp(data1)
# data1 = np.append(data,np.max(data)**2)

mean1 = np.mean(data1)
median1 = np.median(data1)
print(mean1)
print(median1)

y1,x1 = np.histogram(data1,nbins)
xx1 = (x1[1:]+x1[:-1])/2

print(data1)
print(dataexp)
print(' ')
print(xx1)
print('y1:' + str(y1))
print('max y1:' + str(max(y1)))
print('shape y1:' + str(np.shape(y1)))
```

```python
data2 = np.exp(np.random.randn(n))
data2 = np.append(data2,np.max(data2)**6)

mean2 = np.mean(data2)
median2 = np.median(data2)
print(mean2)
print(median2)

y2,x2 = np.histogram(data2,nbins)
xx2 = (x2[1:]+x2[:-1])/2
```

```python
nlarge = 1000

data3 = np.exp(np.random.randn(nlarge))
data3 = np.append(data3,np.max(data3)**2)

mean3 = np.mean(data3)
median3 = np.median(data3)
print(mean3)
print(median3)

y3,x3 = np.histogram(data3,nbins)
xx3 = (x3[1:]+x3[:-1])/2
```

```python
nlarge = 1001

data4 = np.exp(np.random.randn(nlarge))
data4 = np.append(data4,np.max(data4)**6)

mean4 = np.mean(data4)
median4 = np.median(data4)
print(mean4)
print(median4)

y4,x4 = np.histogram(data4,nbins)
xx4 = (x4[1:]+x4[:-1])/2
```

```python
fig,ax = plt.subplots(1,4, figsize=(16,4))

ax[0].plot(xx1,y1,'ks-')
ax[0].plot([mean1,mean1],[0,max(y1)],'r--',label='mean')
ax[0].plot([median1,median1],[0,max(y1)],'b--',label='median')
ax[0].set_title('Small size & small outlier')
ax[0].legend()

ax[1].plot(xx2,y2)
ax[1].plot([mean2,mean2],[0,max(y2)],'r--',label='mean')
ax[1].plot([median2,median2],[0,max(y2)],'b--',label='median')
ax[1].set_title('Small size & large outlier')
ax[1].legend()

ax[2].plot(xx3,y3)
ax[2].plot([mean3,mean3],[0,max(y3)],'r--',label='mean')
ax[2].plot([median3,median3],[0,max(y3)],'b--',label='median')
ax[2].set_title('large size & small outlier')
ax[2].legend()

ax[3].plot(xx4,y4)
ax[3].plot([mean4,mean4],[0,max(y4)],'r--',label='mean')
ax[3].plot([median4,median4],[0,max(y4)],'b--',label='median')
ax[3].set_title('large size & large outlier')
ax[3].legend()

plt.show()
```
