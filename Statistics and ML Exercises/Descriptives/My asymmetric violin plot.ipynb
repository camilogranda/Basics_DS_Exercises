```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
n = 1000

data3 = (np.exp(np.random.normal(loc=0,size=n)))-1
data4 = np.random.normal(loc=1,size=n)
thresh = 6
data3[data3>thresh] = thresh + np.random.rand(sum(data3>thresh))*.1

data = pd.DataFrame({'data1':data3, 'data2':data4})
datam = data.melt()
datam['dummy'] = 0

print(datam)

sns.violinplot(data=datam,y='value',split=True,hue='variable',x='dummy')
plt.show()
```
