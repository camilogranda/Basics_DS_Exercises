```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100

x = np.linspace(0,10,100)
y1 = np.linspace(0,10,100)
y2 = np.logspace(0,10,100)

plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
n = 100

x  = np.linspace(0,10,100)
y1 = np.logspace(0,10,100)
y2 = np.exp(x)

plt.plot(x,y1)
plt.plot(x,y2)

plt.yscale('log')
plt.show()
```

```python
data_lin = np.linspace(start=0.001,stop=10,num=100)
data_log = np.exp(data_lin)

plt.plot(data_log[:40])
plt.plot(data_lin[:40])
plt.show()
```
