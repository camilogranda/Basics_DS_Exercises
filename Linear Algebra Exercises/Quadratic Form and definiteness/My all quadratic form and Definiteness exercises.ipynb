```python
A = np.array([[-2,3],[2,8]])
A

n = 30
wi = np.linspace(-2,2,n)

X,Y = np.meshgrid(wi,wi)

qf = np.zeros((n,n))
qfn = np.zeros((n,n))

for i in range(n):
    for j in range(n):
        W = np.transpose([wi[i],wi[j]])
        
        qf[i,j] = W.T@A@W
        qfn[i,j] = qf[i,j]/(W.T@W)
        
fig = plt.figure(figsize=(5,5))
ax1 = fig.add_subplot(111,projection='3d')
surf1 = ax1.plot_surface(X,Y,qf.T)
ax1.view_init(azim=-30,elev=30)
ax1.set_title('Quadratic form')

fig = plt.figure(figsize=(5,5))
ax2 = fig.add_subplot(111,projection='3d')
surf2 = ax2.plot_surface(X,Y,qfn.T)
ax2.view_init(azim=-30,elev=30)
ax2.set_title('Normalized quadratic form')

plt.show()
```
