```python
import numpy as np
import matplotlib.pyplot as plt
```

### Determinant of shifted matrices

```python
# Shift the matrix 0->.1
idm = np.eye(m,m)
lamb = np.linspace(0,.1,30)
As = np.zeros((m,m))
# Initializa the determinant vector
tmp = np.zeros((1000))
dets = np.zeros((len(lamb)))

for l in range(len(lamb)):
    
    for i in range(1000):
        
        # Generate a square random matrix(20x20 and impose a linear dependence
        m = 20
        A = np.random.randn(m,m)
        A[:,1] = A[:,0]
        
        # Compute the abs(determinant)
        tmp[i] = abs(np.linalg.det(A+lamb[l]*idm))
        
    # Compute average determinant
    dets[l] = np.mean(tmp)
    
plt.plot(lamb,dets,'s-')
plt.xlabel('Fraction of identity for shifting')
plt.ylabel('Determinants')
plt.show()
```
