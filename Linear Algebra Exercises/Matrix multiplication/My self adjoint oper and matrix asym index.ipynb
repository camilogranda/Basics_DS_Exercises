### The matrix asymmetry index

```python
# Create matrix asymmetry index function

def MAI(A):
    Aasym = (A-A.T)/2
    mai = np.linalg.norm(Aasym)/np.linalg.norm(A)
    return mai
```

```python
m = 4

A = np.random.randn(m,m)
print((A-A.T)/2)
print(MAI(A))
```

```python
B = np.array([[2,3,1,3],
              [-2,2,5,3],
              [1,5,3,2],
              [3,-3,2,1]])
Basym = (B-B.T)/2
print(Basym)
print((B-B.T)/2)
print(np.linalg.norm((B-B.T)/2)/np.linalg.norm(B))
```
